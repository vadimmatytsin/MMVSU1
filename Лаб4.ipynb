```python
from scipy.sparse.linalg._dsolve.linsolve import spsolve
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

i = 1
H = 1
k = 0.33
t = np.arange(0,10+H,H)
N=100
def defF(y,t):
  return k*y

t=np.linspace(0, 10 , N)

solve_1=odeint(defF,y0=1,t=t)
solve_1=solve_1.flatten()


def euler(F,y0=1,n=N,h=H):
  eul = np.arange(0,n+h,h)
  solve = [y0]
  for i in range(len(eul)):
    solve.append(solve[i] + h*F(solve[i],eul[i]))
  return solve[:-1], eul

mySolve, eul = euler(defF)
%matplotlib inline 
plt.plot(t, solve_1, label=r"x'=kx", color = "green")
plt.plot(eul, mySolve, label = r"Метод Эйлера", color = "red")
plt.grid(True)
plt.legend()
plt.show()
```
